Importing Library

```python
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

Dataset

```python
dataset = datasets.load_wine()

print('Features: ', dataset.feature_names)
print('Labels: ', dataset.target_names)
```

```python
print(dataset.data.shape)
print(dataset.data[0:5])
print(dataset.target)
```

Membagi Menjadi Data Training dan Data Testing

```python
xTrain, xTest, yTrain, yTest = train_test_split(dataset.data, dataset.target, test_size=0.3, random_state=109)
```

Melakukan Fitting pada Data Training dan Prediksi pada Data Testing

```python
classifier = GaussianNB()
classifier.fit(xTrain, yTrain)

yPred = classifier.predict(xTest)

print(yPred)
print(yTest)
```

Nilai Akurasi Model

```python
print('Accuracy: ', metrics.accuracy_score(yTest, yPred))
```
