Importing Library

```python
#untuk memanipulasi data
import pandas as pd
#untuk grafik
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
#Untuk mengimport dataset bunga iris
from sklearn import datasets
#Untuk KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
```

Memanggil Datasets dan Mengecek Datasets

```python
#memasukkan datasets bunga iris ke dalam variabel
bunga = datasets.load_iris()
#menampilkan tipe objek dari bunga
print(type(bunga))
```

```python
#menampilkan jumlah baris dan kolom dari datasets bunga iris
print(bunga.data.shape)
```

```python
#menampilkan label dari datasets bunga iris
print(bunga.target_names)
```

Mengecek data pada label dan training

```python
#membagi dataset menjadi data label dan data training
x = bunga.data
y = bunga.target

#mengubah tipe dari dataset ke dataframe
df = pd.DataFrame(x, columns=bunga.feature_names)

#menampilkan 5 data pertama dari dataframe
print(df.head())
```

Melakukan Fitting pada Data Training

```python
#memanggil algoritma kNN
knn = KNeighborsClassifier(n_neighbors=6,
                           weights='uniform',
                           algorithm='auto',
                           metric='euclidean')

#fitting model dengan data training dan data targetnya
xTrain = bunga['data']
yTrain = bunga['target']
knn.fit(xTrain, yTrain)
```

Melakukan Prediksi

```python
#data yang akan diprediksi
data = [[6.2, 1.5, 4.2, 2.6]]

#melakukan prediksi berdasarkan 
yPred = knn.predict(data)

#menampilkan hasil prediksi
#jika hasilnya 0, maka Setosa
#jika hasilnya 1, maka Versicolor
#jika hasilnya 2, maka Virginica
print('Hasil Prediksi: ', yPred)
```
